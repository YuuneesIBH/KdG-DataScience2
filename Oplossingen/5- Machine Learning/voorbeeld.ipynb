```python
# üìö Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# üìä Laad de iris dataset
iris = sns.load_dataset('iris')

# üéØ Target & features
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = iris['species']

# üîÄ Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# üå≥ Decision Tree Model
model = DecisionTreeClassifier(max_depth=2, random_state=42)
model.fit(X_train, y_train)

# üîç Voorspellingen
y_pred = model.predict(X_test)

# üìà Evaluatiemetrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')  # Multiclass: average nodig
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# üñ®Ô∏è Resultaten tonen
print(f"Accuracy : {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall   : {recall:.2f}")
print(f"F1 Score : {f1:.2f}")

# üìä Confusion Matrix plotten
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, display_labels=model.classes_)
plt.title("Confusion Matrix")
plt.show()
```
